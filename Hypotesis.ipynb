```python
df = pd.read_csv('metrics.csv')
data =df.transpose()
metrics = data[1][1:6]
recall_model = metrics.to_numpy(dtype=np.float64) 
print(recall_model)

SYSU= pd.read_csv('metrics_SYSU.csv')
data =SYSU.transpose()
metrics = data[1][1:6]
recall_SYSU = metrics.to_numpy(dtype=np.float64)
print(recall_SYSU)

LSTM_PASL= pd.read_csv('metrics_LSTM_PASL.csv')
data =LSTM_PASL.transpose()
metrics = data[1][1:6]
recall_LSTM_PASL = metrics.to_numpy(dtype=np.float64)
print(recall_LSTM_PASL)

SBU = pd.read_csv('metrics_SBU.csv')
data =SBU.transpose()
metrics = data[1][1:6]
recall_SBU = metrics.to_numpy(dtype=np.float64)
print(recall_SBU)

```

```python
from scipy.stats import ttest_ind
import scipy.stats as stats

# Assuming you have the recall values for your model and other models
recall_model = recall_model
recall_other_models = recall_LSTM_PASL

Recall_model_mean = np.mean(recall_model)
Recall_other_models_mean = np.mean(recall_other_models)

recall_model_std = np.std(recall_model)
recall_other_models_std = np.std(recall_other_models)

print("Recall_model_mean:", Recall_model_mean)
print("Recall_other_models_mean:", Recall_other_models_mean)
print("Recall_model_std:", recall_model_std)
print("Recall_other_models_std:", recall_other_models_std)

t = (Recall_model_mean - Recall_other_models_mean)/(recall_model_std/np.sqrt(len(recall_model)))
print("t:", t)

p = 1 - stats.t.cdf(t, len(recall_model)-1)
print("p:", p)



# Perform the t-test
t_statistic, p_value = ttest_ind(recall_model, recall_other_models)

print("t-statistic:", t_statistic)
print("p-value:", p_value)

# Print the results
if p_value < 0.05:
    print("The difference in recall is statistically significant.")
else:
    print("The difference in recall is not statistically significant.")
```
